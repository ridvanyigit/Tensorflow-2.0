```python
url = "https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/Customer_Churn_Modelling.csv"
dataset = pd.read_csv(url)
dataset.head()
```

```python
X = dataset.drop(labels=['CustomerId', 'Surname', 'RowNumber', 'Exited'], axis=1)
y = dataset['Exited']
```

```python
X = pd.get_dummies(X, drop_first=True, columns=['Geography'])
X.head()
```

## Feature Standardization

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

## Build ANN

```python
model = Sequential()
model.add(Dense(X.shape[1], activation='relu', input_dim = X.shape[1]))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
X.shape
```

```python
X.shape[1]
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train.to_numpy(), batch_size = 10, epochs = 10, verbose = 1)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.to_numpy().reshape(len(y_test),1)),1))
```

```python
model.evaluate(X_test, y_test.to_numpy())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm
```
